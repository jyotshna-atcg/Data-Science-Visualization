```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data=pd.read_csv("crcCASt.csv")
data
```

```python
data1=data[data["CRIME HEAD"]=="TOTAL CRIMES AGAINST STs"].loc[380:390,:]
col=list(data1.columns)
col.remove('CRIME HEAD')
data1=data1[col]
data1
```

```python
cur=data1.set_index("STATE/UT").T
plt.rcParams["figure.figsize"] = (10,7)
cur.boxplot()
plt.xticks(rotation=90)
plt.tick_params(axis='x', which='major', labelsize=10)
plt.tight_layout()
plt.xlabel('States of India')
plt.ylabel('Number of crimes committed')
plt.title("Total crimes against STs from 2001-2012 in Indian States",)
plt.show()
```
