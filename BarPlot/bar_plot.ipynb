```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
data=pd.read_csv("NRSC-rainfall-subBasinwise-jan22_1.csv")
#Reference: https://data.gov.in/catalog/rainfall
data
```

```python
data1=data.loc[0:57,:]
data1
```

```python
X_axis= range(1,30)
D1=data1.loc[0:28, 'Avg_rainfall']
D2=data1.loc[29:57, 'Avg_rainfall']
```

```python
plt.rcParams["figure.figsize"] = (10,7)
plt.bar(X_axis, D1, 0.8, label='Area of Inland drainage in Rajasthan Basin')
plt.bar(X_axis, D2, 0.8, label='Barak and Others Basin')
plt.xlabel('January,2022')
plt.ylabel('Average rainfall in millimetre (MM)')
plt.title('Daily rainfall data of Rajasthan & Barak basin during January 2022 ')
plt.grid()
plt.legend(loc=5, bbox_to_anchor=(1.5, 1))
plt.show()

```
