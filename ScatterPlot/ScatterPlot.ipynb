```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data=data=pd.read_csv("rds2011_12_Table_22.csv")
data.shape
data
```

```python
data1=data.loc[0:4, :]
data1.drop(list(data.filter(regex = '%')), axis = 1, inplace = True)
data1
```

```python
X_axis=np.array([1985,1995,2001,2005,2006,2007,2008,2009,2010,2011])
X_axis.shape
np.array(data1.loc[0][2:]).shape

plt.rcParams["figure.figsize"] = (10,7)
for i in data1.index:
    Y=np.array(data1.loc[i][2:])
    plt.scatter(X_axis,Y,label=data1.loc[i][1],alpha=0.5)
    
plt.legend()
plt.title("Faculty wise enrolment in different fields of Higher education from 1985-2011")
plt.tight_layout()
plt.xlabel('Enrolment Year')
plt.ylabel('Enrollment figures in Thousands')
plt.grid()
plt.legend(loc=5, bbox_to_anchor=(1.3, 1))
plt.savefig("ScatterPlot.png",dpi=300)

plt.show()


```
